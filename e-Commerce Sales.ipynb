# E-commerce Sales Data Analysis

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\MAHTAAB AALAM\\Downloads\\ecommerce_sales_34500.csv")
df
```

```python
df.describe()
```

# Drop all unnecessary columns from the datasets

```python
cols = ["order_id","customer_id","product_id"]
df.drop(cols , axis = 1)
```

# Implement bar chart to the categories

```python
category_counts = df["category"].value_counts()

plt.figure(figsize=(6,6))
plt.bar(category_counts.index , category_counts.values , color = ["Red","Violet","Black","Teal","Purple","Blue","Yellow"])
plt.xlabel("Category")
plt.ylabel("Counts")
plt.title("Top Selling Category of the Sales")
plt.tight_layout()
plt.show()
```

# Implement barchart to show quantity 

```python
quantity_counts = df["quantity"].value_counts()

plt.figure(figsize=(6,6))
plt.bar(quantity_counts.index , quantity_counts.values , color = ["Red","Violet","Black","Teal","Purple"])
plt.xlabel("Quantity")
plt.ylabel("Counts")
plt.title("Frequency of the Product Quantity")
plt.tight_layout()
plt.show()
```

# Implement pie-chart to show payment methods 

```python
payment_method_counts = df["payment_method"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(payment_method_counts, 
        labels=payment_method_counts.index, 
        autopct="%0.2f%%", 
        colors=["Red", "Violet", "Teal", "Purple", "Blue", "Yellow"])

plt.title("Percentage of Payment Methods")
plt.tight_layout()
plt.legend()
plt.show()
```

# Implement the bar plot to show the delivery time

```python

delivery_df = df.copy()

plt.figure(figsize=(6, 6))
plt.hist(delivery_df["delivery_time_days"], color="Black", edgecolor="Red")

plt.xlabel("Delivery Time (days)")
plt.ylabel("Number of Orders")
plt.title("Distribution of Delivery Time")
plt.tight_layout()
plt.show()
```

# Implement the harizontal bar plot to show the region

```python
region_counts = df["region"].value_counts()

plt.figure(figsize=(6,6))
plt.barh(region_counts.index , region_counts.values , color = ["Black"])
plt.xlabel("Counts")
plt.ylabel("Regional Sales")
plt.title("Top Sales Across the Region")
plt.tight_layout()
plt.show()
```

# Implement Histogram plot to show the cutomer's age

```python
plt.figure(figsize=(6, 6))
plt.hist(df["customer_age"], color="Red", edgecolor="Black", bins=20)

plt.xlabel("Customer's Age")
plt.ylabel("Number of Buyers")
plt.title("Age Distribution of Buyers")
plt.tight_layout()
plt.show()
```

# Implement pie - chart to show gender of the buyers

```python
customer_gender_counts = df["customer_gender"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(customer_gender_counts, 
        labels=customer_gender_counts.index, 
        autopct="%0.2f%%", 
        colors=["Red","Purple","Teal"])

plt.title("Percentage of Customer's Gender")
plt.tight_layout()
plt.legend()
plt.show()
```
